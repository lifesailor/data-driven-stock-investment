# 1. 나노

```python
df = krx.get_market_ohlcv("20140101", "20190306", "187790")
df.index = pd.to_datetime(df.index)
minimum = min(df['종가'])
maximum = max(df['종가'])

plt.figure(figsize=(12,5))
df['종가'].plot()
plt.title("Total")
plt.ylim([minimum, maximum])
plt.show()

## 봄 효과
years = ['2015','2016','2017','2018']
for year in years:
    plt.figure(figsize=(12,5))
    df_sub = df[year + '0101': year + '0630']
    minimum = min(df_sub['종가'])
    maximum = max(df_sub['종가'])
    df.loc[year + '0101': year + '0630']['종가'].plot()
    plt.xticks(rotation=90)
    plt.title(year)
    plt.ylim([minimum, maximum])
    plt.show()

year = '2019'
plt.figure(figsize=(4,5))
df_sub = df[year + '0101': year + '0630']
minimum = min(df_sub['종가'])
maximum = max(df_sub['종가'])
df.loc['20190101':'20190630']['종가'].plot()
plt.xticks(rotation=90)
plt.title("2019")
plt.ylim([minimum, maximum])
plt.axvline(x=pd.to_datetime('20190301'), c='r', ymax=maximum, ymin=minimum)
plt.show()
```

```python
years = ['2014', '2015', '2016', '2017', '2018', '2019']

for i, year in enumerate(years):
    print(years[i+1] + " 1월 1일 시가: ", df[years[i] + '-12-31':].iloc[0]['시가'])
    print(years[i+1] + " 3월 4일 종가: ", df[:years[i+1]  + '-03-05'].iloc[-1]['종가'])
    print(years[i+1] + " 1월 1일 - 3월 4일 종가 시가 차이", (df[:years[i+1]  + '-03-05'].iloc[-1]['종가'] - df[years[i] + '-12-31':].iloc[0]['시가']))
    print(years[i+1] + " 1월 1일 - 3월 4일 고점 저점 비율", (df[:years[i+1]  + '-03-05'].iloc[-1]['종가'] - df[years[i] + '-12-31':].iloc[0]['시가']) / (df[year + '-12-31':].iloc[0]['시가']))
    print("\n")
    
    if i == len(years) - 2:
        break
```

```python
years = ['2015', '2016', '2017', '2018', '2019']

for i, year in enumerate(years):
    print(years[i] + " 3 - 6월 고점 ", (np.argmax(df[years[i] + '-03-01':years[i]  + '-06-30']['고가'])))
    print(years[i] + " 3 - 6월 고점 ", (max(df[years[i] + '-03-01':years[i]  + '-06-30']['고가'])))
    print(years[i] + " 3 - 6월 고점 ", (np.argmin(df[years[i] + '-03-01':years[i]  + '-06-30']['고가'])))
    print(years[i] + " 3 - 6월 저점 ", (min(df[years[i] + '-03-01':years[i]  + '-06-30']['저가'])))
    print('\n')
        
    if i == len(years) - 2:
        break
```
