```python
import datetime as dt
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 1. Import Test
from pykrx import Krx
```

```python
krx = Krx()
```

# 1. 나노

```python
df = krx.get_market_ohlcv("20140101", "20190304", "187790")
df.index = pd.to_datetime(df.index)
minimum = min(df['종가'])
maximum = max(df['종가'])

plt.figure(figsize=(12,5))
df['종가'].plot(c='g')
plt.title("Total")
plt.ylim([minimum, maximum])
plt.show()

## 봄 효과
years = ['2015','2016','2017','2018']
for year in years:
    plt.figure(figsize=(12,5))
    df_sub = df[year + '0701': year + '1231']
    minimum = min(df_sub['종가'])
    maximum = max(df_sub['종가'])
    df.loc[year + '0701': year + '1231']['종가'].plot(c='g')
    plt.xticks(rotation=90)
    plt.title(year)
    plt.ylim([minimum, maximum])
    plt.show()
```
